```python
portfolio = [
    {"name": "Bitcoin", "symbol": "BTC", "quantity": 1.5, "purchase_price": 50000},
    {"name": "Ethereum", "symbol": "ETH", "quantity": 5, "purchase_price": 3000},
    {"name": "Ripple", "symbol": "XRP", "quantity": 1000, "purchase_price": 1.5},
    {"name": "Litecoin", "symbol": "LTC", "quantity": 10, "purchase_price": 200},
    {"name": "Cardano", "symbol": "ADA", "quantity": 500, "purchase_price": 2},
    {"name": "Polkadot", "symbol": "DOT", "quantity": 20, "purchase_price": 40},
    {"name": "Chainlink", "symbol": "LINK", "quantity": 50, "purchase_price": 25},
    {"name": "Stellar", "symbol": "XLM", "quantity": 2000, "purchase_price": 0.4},
]

def calculate_portfolio_value(portfolio):
    total_value = 0
    for crypto in portfolio:
        total_value += crypto['quantity'] * crypto['purchase_price']
    return total_value

def select_cryptocurrency():
    print("Select a cryptocurrency:")
    for i, crypto in enumerate(portfolio, 1):
        print(f"{i}. {crypto['name']}")

    while True:
        choice = input("Enter the number of the cryptocurrency you want to select (or 'q' to quit): ")
        if choice.lower() == 'q':
            return None
        elif choice.isdigit():
            choice_idx = int(choice) - 1
            if 0 <= choice_idx < len(portfolio):
                return portfolio[choice_idx]
            else:
                print("Invalid selection. Please choose a valid number.")
        else:
            print("Invalid input. Please enter a number.")

def main():
    while True:
        selected_crypto = select_cryptocurrency()
        if selected_crypto is None:
            print("Exiting.")
            break

        print("\nSelected Cryptocurrency Details:")
        print(f"Name: {selected_crypto['name']}")
        print(f"Quantity: {selected_crypto['quantity']}")
        print(f"Purchase Price: ${selected_crypto['purchase_price']}\n")

    total_value = calculate_portfolio_value(portfolio)
    print(f"\nTotal Portfolio Value: ${total_value}")

if __name__ == "__main__":
    main()
3
```
